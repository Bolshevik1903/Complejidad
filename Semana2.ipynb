##Selection Sort

```python
def SelectionSort(lista):
  n = len(lista)
  for i in range(n-1):
    min_pos = i
    for j in range(i+1, n):
      if lista[j] < lista[min_pos]:
        min_pos = j

    #Hacer la permutación en el caso que la variable min_pos != i
    if min_pos != i:
      lista[min_pos], lista[i] = lista[i], lista[min_pos]
```

##Bubble Sort

```python
def BubbleSort(lista):
  n = len(lista)
  for i in range(n-1):
    for j in range(n-1-i):
      if lista[j] > lista[j+1]:
        lista[j], lista[j+1] = lista[j+1], lista[j]
```

# Backtracking (Volver atras)
<ul>
  <li> Busca sistemáticamente a través de todas las configuraciones de un espacio de búsqueda.
  <li> Dada una solución S, se procede a </li>
    <ol>
      <li>Verificar si S es solución. Si lo es, hacen algo con ella
(depende del problema).
      <li>Construyen todas las posibles extensiones de S, e
invocan recursivamente al algoritmo con todas ellas.
    </ol>
</ul>

##Ejemplos
<ul>
  <li> 4 Reinas
  <li> Viajero
</ul>

##Ventajas
<ul>
  <li> Resuelve casi cualquier problema
  <li> Utilizado para todas las soluciones existentes
  <li> Representación de solución paso a paso
  <li> Muy facil en código y de depurar
</ul>

##Desventajas
<ul>
  <li> Muy lento comparado con otras soluciones
  <li> Posible falla al tener muchos datos
  <li> Costo computacional alto
</ul>

```python
#Problema de las 9 reinas
import numpy as np #Para trabajar con arreglos
import matplotlib.pyplot as plt #Para trabajar con gráficos

tablero = [4,7,2,6,1,0,3,5]

for fil, col in enumerate(tablero):
  print(f"Hay una reina en la fila: {fil} y en la columna: {col}")
```

```python
#Funcion para dibujar un tablero

def draw(tablero):
  n = len(tablero)
  #np.zeros devuelve un array del tamaño y tipo indicados inicializando sus valores con ceros
  b = np.zeros((n, n, 3), dtype=int)
  #print(b)
  b += [255, 128, 80]

  b[::2, ::2] = [255, 225, 120]
  b[1::2, 1::2] = [255, 225, 120]

  _, ax = plt.subplots()
  ax.imshow(b)
  for fil, col in enumerate(tablero):
    ax.text(col, fil, u"\u265b", fontsize=200/n, va="center", ha="center")
  ax.set(xticks=[], yticks=[])
  '''
  La función set(xticks=[], yticks=[]) toma un objeto de lista como argumento. Los elementos de la lista indican las posiciones en la acción correspondiente donde se mostrarán las marcas. 
  Este método marcará los puntos de datos en las posiciones dadas con marcas.
  '''

draw(tablero)
```

```python
#Función para validar tablero
def validar(tablero, fil, col):
  for fil_i in range(fil):
    col_i = tablero[fil_i]
    delta = fil - fil_i #Fila anterior
    if col in (col_i, col_i + delta, col_i - delta): #Posiciones no validas de las coranas en la columna
      return False
    
  return True
```

```python
tablero1 = [1,3,-1,-1]
draw(tablero1)
```

```python
#Podemos colocar una reina en la posicion: fila 2 y columna 0?
#Función assert para controlar la respuesta que da la función
assert validar(tablero1, 2, 1) == False
```

```python
#Generalizar la funcion para N reinas
def nqueens(tablero, fil):
  n = len(tablero)
  if fil == n:
    #dibujar el trablero cuando se evalua la ultima fila
    draw(tablero)
  else:
    for col in range(n):
      if validar(tablero, fil, col):
        tablero[fil] = col
        #llamada recursiva para evaluar la siguiente fila del tablero hasta la última fila
        nqueens(tablero, fil+1)
```

```python
n = 4
tablero2 = [-1]*n
print(tablero2)
```

```python
nqueens(tablero2, 0)
```
